# PLOT

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
x=[12,43,23,41,26]
y=[41,21,12,19,20]

plt.plot(x,y, marker = ".",ms = "12", mec= "red", mfc = "blue",linestyle="dashdot")
plt.xlabel ("This is X axis", color = "orange")
plt.ylabel ("Y axis is here", color= "blue")
plt.title ("Title is here", loc="left", color = "Green")
plt.grid()

for i in range (len(x)):
    plt.annotate(xy=(x[i],y[i]),text = f"({x[i]},{y[i]})",color = "brown")

```

# SCATTER PLOT

```python
x = np.array ([12,32,13,32,25,16])
y = np.array ([19,23,52,14,52,26])


# x = random.rand(1,10)
# y = random.rand (1,10)
for i in range (len(x)):
    plt.annotate (xy=(x[i],y[i]),text = f"({x[i]},{y[i]})",color = "red")


plt.xlabel("Group A")
plt.ylabel("Group B")
plt.scatter (x,y)

# plt.legend()
```

```python
age = np.array([23,28,21,25,30,32,29,33])
investment = np.array([2500,5100,1200,2400,4300,1300,4210,3700])
income = [120,230,320,420,190,299,150,810]


# income.shape
for i in range (len(age)):
    plt.annotate (xy=(age[i],investment[i]),text = f"({age[i]},{investment[i]})", color = "green")



plt.scatter(age,investment, s=income)
```

# PIE CHART

```python
data = [10,20,19,30,25,35]

name = ["Apple","Grapes", "Kiwi", "Mango", "Watermelon", "Banana"]

# plt.pie(data, labels=name)
clr = ["Red","Green", "purple", "Yellow", "Grey", "orange"]
value= [0.1,0,0,0,0,0.2]
plt.pie (data ,labels = name, explode = value, colors=clr) 
# Use Labels fun- to mention the fruit/lables name, explode= To highlight the data, Colors = to assign colors.

plt.legend(loc = "lower left") # To describe the elements of the graph.
```

```python
#### import random
# import plotly.graph_objects as go

# Generate random data
data1 = random.randint(2, 50, size=14)
data2 = random.randint(30, 60, size=14)
data3 = random.randint(2, 50, size=14)

# Create the 3D scatter plot
fig = go.Figure(data=go.Scatter3d(x=data1, y=data2, z=data3, mode='markers'))

# Customize the plot (optional)
fig.update_layout(
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    ),
    title='Random 3D Scatter Plot'
)

# Show the plot
fig.show()
```

```python
#USING MODE PARAMETER
import plotly.graph_objects as go


x = [12,42,22,41,34,53]
y = [32,51,13,16,25,15]
fig=go.Figure(
    # data = go.Scatter(x=x, y=y ,mode = "lines")   #lines', 'markers', 'text'] joined with '+' characters
    #     data = go.Scatter(x=x, y=y ,mode = "markers")  
        # data = go.Scatter(x=x, y=y ,mode = "text")
        data = go.Scatter(x=x, y=y ,
                         mode = "markers+lines",
                         marker= dict (color='red',size = 12),
                         hoverinfo= "x+y",
                         hovertext= "Defined with the help of Hovertext",
                         line = dict (color = "orange", width = 4, shape= "spline")

                         # To shape line we can also use "dash"= "dot/dot/solid/longdashdot"                           
                        # for the SHAPE property - ['linear', 'spline', 'hv', 'vh', 'hvh', 'vhv']
) 
)
# fig.add_annotation(x=x[3],y=y[3],text= "Hello",showarrow=True) 
for i in range (len(x)):
    fig.add_annotation(x=x[i],y=y[i],text = f"({x[i]},{y[i]})")
    

fig.show()


# WAP to show the coordinates without hovering at the coordinates.





```
