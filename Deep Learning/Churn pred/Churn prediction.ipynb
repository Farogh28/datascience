```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Churn_Modelling.csv")
```

```python
data
```

```python
data.drop(columns = ["RowNumber", "CustomerId","Surname"], inplace = True)
```

```python
data
```

```python
# Converting categorical data into numerical one
# with the help of pandas funcion get_dummies (similar with one hot encoder)

change = pd.get_dummies(data, columns = ["Geography", "Gender"],drop_first= True)  
change
```

```python
# Now we are changing the data type of the newly generated data

new_data = change.astype(int)
new_data
```

```python
# Seprating data

x = new_data.drop(columns = ["Exited"])
x
```

```python
y = new_data["Exited"]
y
```

```python
y[2]
```

```python
# data spliting 

from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
y_train.shape
```

```python
# Norm alize the data (via Standard scaler)

from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
sc
```

```python
x_train_new = sc.fit_transform(x_train)
x_train_new
```

```python
import tensorflow

from tensorflow import keras
from tensorflow.keras.optimizers import Adam
```

```python
model  = tensorflow.keras.Sequential()
model
```

```python
model.add(tensorflow.keras.layers.Dense(11, activation = 'sigmoid', input_dim = 11))  # activation = Softmax
model.add(tensorflow.keras.layers.Dense(11, activation = 'sigmoid'))
model.add(tensorflow.keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
model
```

```python
model.compile(optimizer = "Adam", loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.summary()
```

```python
model.fit(x_train_new, y_train, epochs =10)
```

```python
y_pred = model.predict(x_train_new)
y_pred
```
