# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
x= np.random.rand(80, 1)
y= 4 + 5 * x + np.random.randn(80, 1)
# y= 4 + 5 * x + 0.2 * np.random.randn(80, 1)
```

```python
# y , x
```

```python
reg= LinearRegression()
```

```python
reg.fit (x,y)
```

```python
plt.scatter(x,y)
```

```python
x_vals = np.linspace(0,1,80).reshape(-1, 1)
y_vals = reg.predict(x_vals, y_vals, color = 'red')
plt.show()
```
