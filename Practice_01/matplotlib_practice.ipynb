```python
from matplotlib import pylab
print (pylab.__version__)   # to know the version
```

```python
import numpy as np


x = np.linspace(0,10,25)
y = x*x+2

print (x)
print (y)

# print (np.array([x,y]).reshape(25,2))   # Reshape like this to plot properly
```

```python
#  It only takes 1 command to draw

pylab.plot(x,y, 'r')    # r stands for red 
```

```python
#  Draw a subgraph


pylab.subplot(1,2,1)       # the context of the brackets represents {rows, columns, indexes}
pylab.plot(x,y,'r--')      # r-- is for color and line style
pylab.subplot(1,2,2)   #  Subplot (row, column, index)
pylab.plot(x,y, 'g*-')
```

```python
pylab.subplot(1,2,1) 
pylab.plot(x,y, 'r--')

pylab.subplot(2,1,2) 
pylab.plot(x,y, 'g*-')
```
