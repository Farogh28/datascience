```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta  = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x,y)
z1 = np.exp(-X**2 - Y**2)
z2 = np.exp(-(X -1)**2 - (Y-1) **2)
z = (z1-z2)*2

# print (x)
# print(Y)
```

```python
fig, ax = plt.subplots()
cs = ax.contour(X,Y,z)
ax.clabel(cs, inline = 1, fontsize = 10)
ax.set_title('Contour Plot')
```
