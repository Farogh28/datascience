```python
from matplotlib import pylab
print (pylab.__version__)   # to know the version
```

```python
import numpy as np


x = np.linspace(0,10,25)
y = x*x+2

print (x)
print (y)

# print (np.array([x,y]).reshape(25,2))   # Reshape like this to plot properly
```

```python
#  It only takes 1 command to draw

pylab.plot(x,y, 'r')    # r stands for red 
```

```python
#  Draw a subgraph


pylab.subplot(1,2,1)       # the context of the brackets represents {rows, columns, indexes}
pylab.plot(x,y,'r--')      # r-- is for color and line style
pylab.subplot(1,2,2)   #  Subplot (row, column, index)
pylab.plot(x,y, 'g*-')
```

```python
pylab.subplot(1,2,1) 
pylab.plot(x,y, 'r--')

pylab.subplot(2,1,2) 
pylab.plot(x,y, 'g*-')
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()

axis = fig.add_axes([0.5, 0.1, 0.8,0.8])  # control the left, width, height of the canvas(from 0 to 1)

axis.plot (x,y, 'r')
```

```python
#  plot subgraph

fig, axes = plt.subplots(nrows = 1, ncols = 2)
for ax in axes:
    ax.plot(x,y, 'r')
```

```python

fig, axes = plt.subplots(nrows = 2, ncols = 1)
for ax in axes:
    ax.plot(x,y, 'r')
```

```python

fig, axes = plt.subplots(nrows = 1, ncols = 3)
for ax in axes:
    ax.plot(x,y, 'r')
```

```python
# draw a picture/ graph inside another graph

# Controls:  (LEFT, BOTTOM, WIDTH, HEIGHT OF Canvas (scale from 0 to 1))

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])   # Big Axes    (0.1,0.8 --> these are the coordinates on graph)
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])   # Small canvas

axes1.plot(x,y, 'r')
axes2.plot (y,x, 'g')
```
