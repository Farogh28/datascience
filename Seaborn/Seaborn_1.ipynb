```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


```

```python
a = [1,2,3,4,5,6,7,8]
c = [2,3,4,5,3,3,5,6]

df = pd.DataFrame({"a": a, "c" : c})

# or 

df = pd.DataFrame({"a": [1,2,3,4,5,6,7,8], "c": [2,3,4,5,3,3,5,6]})


sns.lineplot(x = "a", y= "c", data = df)

plt.show()   # To remove title  from graph
```
