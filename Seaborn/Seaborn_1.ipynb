```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


```

```python
fmri = sns.load_dataset("fmri")   # fmri is builtin dataset
fmri.head()
# fmri.tail()
```

```python
sns.lineplot(x = "timepoint", y= "signal", data = fmri)
plt.show()
```

```python
# HUE - Visualize data in one plot
```

```python
sns.lineplot(x = "timepoint", y= "signal", data = fmri, hue = "event")
plt.show()
```

```python
# STYLE a line 
```

```python
sns.lineplot(x = "timepoint", y= "signal", data = fmri, hue = "event", style = "event")
plt.show()
```

```python
# MARKERS --  To add markers on the line 
```

```python
sns.lineplot(x = "timepoint", y= "signal", data = fmri, hue = "event", style = "event", markers = True)
plt.show()
```

# HISTOGRAM or a DISTRIBUTION PLOT

```python
diamond = pd.read_csv("diamonds.csv")
```

```python
diamond.head()
```

```python
# Distribution plot - is used for continous numerical values,
```

```python
sns.distplot(diamond['price'])
plt.show()
```

```python
""" if want only frequency curve without histogram   {hist = False}"""
```

```python
sns.distplot(diamond['price'], hist = False)
plt.show()
```

# Color

```python
sns.distplot(diamond['price'], color = "green")
plt.show()
```

```python
# Dist plot without frequency curve
```

```python
""" KDE = False """
```

```python
sns.distplot(diamond['price'], kde = False )
plt.show()
```

```python
sns.distplot(diamond['price'], kde = False, bins = 10 )
plt.show()
```

```python
sns.distplot(diamond['price'], kde = False , vertical= True)
plt.show()
```
