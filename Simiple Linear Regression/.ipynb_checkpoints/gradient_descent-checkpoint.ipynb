```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv("C:\\Users\\Farogh\\Desktop\\learning\\placement.csv")
x = data["cgpa"]
y = data["package"]
# x = np.array([1, 2, 3, 4, 5])
# y = np.array([5, 9, 7, 10, 11])
```

```python
def gradient_descent(x, y):
    m_curr = b_curr = 0
    iterations = 20000
    n = len(x)
    learning_rate = 0.01

    for i in range(iterations):
        y_pred = m_curr * x + b_curr
        cost = (1/n) * sum([val** 2 for val in (y - y_pred)])
        # y = mx+b

        # df/dm, df/db

        m_der = (-2/n) * sum(x * (y - y_pred))
        b_der = (-2/n) * sum((y - y_pred))

        # adjusting values
        m_curr = m_curr - learning_rate * m_der
        b_curr = b_curr - learning_rate * b_der
        
        print('m = {}, b = {}, cost = {} iteration = {}'.format(m_curr, b_curr,cost, i))
```

```python
gradient_descent(x, y)
# m_curr = 2.021391068947519,
# b_curr = 2.9227714537500327
```

```python
plt.scatter(x, y, color = 'red')
plt.plot(x, m_curr * x + b_curr, color = 'blue')
plt.show()
```
