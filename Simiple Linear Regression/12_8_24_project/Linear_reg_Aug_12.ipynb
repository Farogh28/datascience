```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("linear regression dataset.csv")
data.head(4)
```

```python
data.shape
```

```python
data.info()
```

```python
data = data.dropna()
data.shape
```

```python
x = data.iloc[:, :-1]
x
```

```python
y= data.iloc[:,-1]
y
```

```python
plt.scatter (x,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
print (f"x_train shape: {x_train.shape}, {"\n"}y_train shape {y_train.shape}")
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(x_train, y_train)
```

```python
m = model.coef_
```

```python
b = model.intercept_
```

```python
# y = mx+b
```

```python
y = m * 24 + b
y
```

```python
y_pred = model.predict(x_test)
```
