```python
df = pd.DataFrame({
    "Age" : [12,34,23,45,23,34,23,45,23],
    "Fav_color" : ["blue", "red", "green", "red", "pink", "black", "pink", "blue", "blue"],
    "number" : [234345345,235456454,74564353,754678463,856345346,25464577458,8654654443,475745684534,4564574578]
})
```

```python
df
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
OHE_color = ohe.fit_transform(df[["Fav_color"]])
```

```python
OHE_color
```

```python
OHE_color.toarray()    # To convert sparse matrix into array format
```

```python
df
```

```python
data = pd.DataFrame(OHE_color.toarray())    # to add values in DF
```

```python
data
```

```python
data = pd.DataFrame(OHE_color.toarray(), columns= ohe.get_feature_names_out(["Fav_color"]))   # To show which color is assign to which category
```

```python
data
```

```python
xyz = pd.concat([df, data], axis= 1)   #To concatinate two data frames **************
```
