```python
# WCSS applying
```

```python
from sklearn.cluster import KMeans # k-mean is an algorithm so K and M is inn capital form
```

```python
# WCSS calculate to know how many cluster is going to form
```

```python
WCSS = []
```

```python
# for i in range (1,11):
km = KMeans(n_clusters = 1)    #(paramenter is used in k means is n_cluster )

km.fit_predict(data)

km.inertia_   #inertia is a paramnter or a wcss
```

```python
km = KMeans(n_clusters = 2)    #(paramenter is used in k means is n_cluster )

km.fit_predict(data)

km.inertia_   #inertia is a paramnter or a wcss
```

```python
km = KMeans(n_clusters = 3)    #(paramenter is used in k means is n_cluster )

km.fit_predict(data)

km.inertia_   #inertia is a paramnter or a wcss
```

```python
km = KMeans(n_clusters = 4)    #(paramenter is used in k means is n_cluster )

km.fit_predict(data)

km.inertia_   #inertia is a paramnter or a wcss
```

```python
km = KMeans(n_clusters = 5)    #(paramenter is used in k means is n_cluster )

km.fit_predict(data)

km.inertia_   #inertia is a paramnter or a wcss
```

```python
# back to the loop
```

```python
WCSS = []
```

```python
for i in range (1,11):
    km = KMeans(n_clusters = i)    
    km.fit_predict(data)
    WCSS.append(km.inertia_)  
```

```python
WCSS
```

```python
plt.plot (range(1,11), WCSS, color ="red", linewidth = 2, label = "Elbow curve", 
          marker = ".", ms = "10", mfc = "yellow", mec = "green")
plt.legend()
    
for i in range (1,5):        # to print, to show after how many cluster the value is coming
    plt.annotate(xy = (i, WCSS[i-1]), text = f"Centroid : {i}, WCSS: {WCSS[i-1]}")
```
